```python
#seperating the data
c_c_data_legit = c_c_data[c_c_data.Class == 0]
c_c_data_fraud = c_c_data[c_c_data.Class == 1]
```

```python
c_c_data_legit_sample = c_c_data_legit.sample(n=102)
```

```python
new_dataset = pd.concat([c_c_data_legit_sample, c_c_data_fraud], axis=0)
```

```python
from re import X
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
X_train, X_text,Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_text.shape)
```

```python
model = LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
X_text_prediction = model.predict(X_text)
text_data_accuracy = accuracy_score(X_text_prediction, Y_test)
```
